# Zad6

```python
cabin = {}
for index,row in enumerate(set(data["cabin"])):
    cabin[row] = index

CabinReduced = {}
for index,row in enumerate(set(data["CabinReduced"])):
    CabinReduced[row] = index

sex = {}
for index,row in enumerate(set(data["sex"])):
    sex[row] = index   
```

# Zad7

```python
data["cabin_map"] = data["cabin"].map(cabin)
data["CabinReduced_map"] = data["CabinReduced"].map(CabinReduced)
data["sex_map"] = data["sex"].map(sex)
```

```python
col_name = ["cabin","cabin_map","CabinReduced","CabinReduced_map","sex","sex_map"]

X_train, X_test, y_train, y_test = train_test_split(data[col_name], data["survived"], test_size=0.3, random_state=0)
```

# Zad8

```python
print("X_train",pd.isnull(X_train).sum(),sep='\n')
print("X_test",pd.isnull(X_test).sum(),sep='\n')
print("y_train",pd.isnull(y_train).sum(),sep='\n')
print("y_test",pd.isnull(y_test).sum(),sep='\n')
```

Brakujące wartości występują jedynie dla zmiennej "cabin" i analogicznie w "CabinReduced". Dla zbioru testowego jest to 30.8 % całości brakujacych wartosci, czyli niemal tyle samo co jego stosunek do całosci zbioru.

```python
X_train = X_train.replace(np.nan,0)
X_test = X_test.replace(np.nan,0)
print("X_train",pd.isnull(X_train).sum(),sep='\n')
print("X_test",pd.isnull(X_test).sum(),sep='\n')
```

Podczas zamiany tych zmiennych na zmienne numeryczne, wartosci NaN zostały juz przypisane do wartosci "0". Uważam, że lepiej skorzystać z tych wartosci zamiast zamieniac w poprzednich Nan na wartosc liczbową 0 podczas gdy pozostałe wartosci są stringami.
