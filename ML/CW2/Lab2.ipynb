W pierwszych dwudziestu wierszach wyróżnia się duża ilość brakujących danych w kolumnie "body" oraz niecałkowita wartość wieku w 2 kolumnie. Ponadto brakujące dane są zapisane w postaci str '?'.

```python
pd.isnull(data).sum() 
pd.isnull(data).mean()
```

Przed zastosowaniem powyższych funkcji należy zamienić znaki '?' na wartosci logiczne np. None

```python
data = data.replace('?',np.nan) 
pd.isnull(data).sum() 
```

```python
pd.isnull(data).mean()
```

Zakłada się, że usunięcie danych brakujących w ilości < 5% całości zbioru jest akceptowalne. W naszym przypadku taką operacje możemy wykonać dla 'fare' oraz 'embarked'. W przypadku pozostałych brakujących danych np. 'body' ( 91 % ), należy zastosować inne metody.

```python
for col in data.columns:
    if pd.isnull(data[col]).sum() != 0:
        data[col+'Null'] = np.where(data[col].isnull(), 1, 0); print(data.groupby(['survived'])[col+'Null'].mean())
        
```

W przypadku brakujacych danych dla kolumn 'cabin', boat' oraz 'body' występuje silna zależność z kolumną 'survived' co oznacza że te dane możemy sklasyfikowac jako MNAR. 
W przypadku brakujacych danych dla kolumn 'age' oraz 'home.dest' zależność ta jest mniejsza natomiast nadal powinnismy je sklasyfikowac jako MNAR.
W przypadku brakujacych danych dla kolumn 'fare' oraz 'embarked' zależność nie występuje i możemy je sklasyfikowac jako MAR lub MCAR.

Zapis w jednej lini kodu ( do zrobienia )
